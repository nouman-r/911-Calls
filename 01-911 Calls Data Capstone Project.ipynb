## Creating new features

```python
df['Reason'] = df['title'].apply(lambda x: x.split(':')[0])
```

** What is the most common Reason for a 911 call based off of this new column? **

```python
df['Reason'].value_counts()
```

```python
#df['Reason'].value_counts().plot(kind= 'bar')
sns.countplot(data= df, x= 'Reason')
```

```python
df['timeStamp'].apply(lambda x: 'STR' if type(x)!= 'str' else 'NOT STR').value_counts() # all are STR
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df['Hour'] = df['timeStamp'].apply(lambda x: x.hour)
df['Month'] = df['timeStamp'].apply(lambda x: x.month)
df['Day of Week'] = df['timeStamp'].apply(lambda x: x.dayofweek)
df.head()
```

** Notice how the Day of Week is an integer 0-6. Use the .map() with this dictionary to map the actual string names to the day of the week: **

    dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day of Week'] = df['Day of Week'].apply(lambda x: dmap[x])
```

```python
plt.style.use('ggplot')
sns.countplot(data=df, x='Day of Week', hue = 'Reason', palette='viridis')
plt.legend(loc = (1,0))
```

```python
sns.countplot(data=df, x='Month', hue = 'Reason', palette='viridis')
plt.legend(loc = (1,0))
```

```python
df['Date'] = df['timeStamp'].apply(lambda x: x.date())
```

```python
df.groupby(by='Date').count()
```

** Now recreate this plot but create 3 separate plots with each plot representing a Reason for the 911 call**

```python
print(plt.style.available)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(6,4),frameon=True)
df[df['Reason'] == 'Traffic'].groupby(by='Date').count()['twp'].plot(c='blue', title='Traffic')
plt.tight_layout()
```

```python
df[df['Reason'] == 'Fire'].groupby(by='Date').count()['twp'].plot(c='green', title='Fire', figsize=(6,4))
plt.tight_layout()
```

```python
df[df['Reason'] == 'EMS'].groupby(by='Date').count()['twp'].plot(title='EMS',figsize=(6,4))
plt.tight_layout()
```

```python
#df.pivot_table(columns='Hour', index='Day of Week', values='Hour')
byDayofWeek = df[['Day of Week', 'Hour', 'e']].pivot_table(columns='Hour', index='Day of Week', values='e', aggfunc=np.sum)
```

** Now create a HeatMap using this new DataFrame. **

```python
sns.heatmap(data=byDayofWeek, cmap='viridis')
```

** Now create a clustermap using this DataFrame. **

```python
sns.clustermap(byDayofWeek, cmap='viridis')
```
